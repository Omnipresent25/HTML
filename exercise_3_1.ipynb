```python
import pandas as pd
url = "https://data.cdc.gov/api/views/v6ab-adf5/rows.csv?accessType=DOWNLOAD"
mortality_wide = pd.read_csv(url)
```

```python
mortality_wide.to_pickle('mortality_Year.pkl.column')
```

```python
#using chaining to get the results.
```

```python
mortality_wide.plot.line(title='Child Mortality: 15-19 [AgeGroup], DeathRate:01-19')
```

```python
mortality_wide.plot.area(
    title=['Child Mortality: 01-04','Child Mortality:15-19'],
    legend=False, subplots=True, layout=(2,2), figsize=(10,10))
```

```python
mortality_wide.query('Year in [1900]').plot.barh()
```

```python
mortality_wide.query('Year in [2000]').plot.barh(
    title=['Child Mortality: 10-14','Child Mortality: 15-19'], subplots=True, layout=(2,2),figsize=(10,10))
```

```python
mortality_wide.query('Year in (1900,2000)').plot.barh(
    title=['Child Mortality: 10-14','Child Mortality: 15-19'], subplots=True, layout=(2,2),figsize=(10,10))
```

```python
mortality_wide.plot.hist(y=(), bins=15)
```

```python
mortality_wide.plot.density(y=())
```

```python
mortality_wide.plot.box()
```

```python
mortality_wide.plot.line(
    title=['Child Mortality: 01-04', 'Child Mortality: 05-09'],
           ylabel='Deaths per 100,000', grid=True, rot=45, xlim=(1900-2000), legend=False,
           subplots=True, layout=(2,2), figsize=(10,10))
```
